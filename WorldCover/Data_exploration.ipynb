### 3. Answers to questions

##### 3.1. Do crop yields (in kg/acre) vary significantly from region to region, from district to district, from farmer to farmer? What seems to be a normal, vs. low, vs. high yield for maize in Ghana?

There are a couple of fields in cryield table that indicates crop yields. I think that the best indicator is 'yield_backs' and 'planted acres'. In the end, we should multiply 'yield_backs' on 100 to get data in kg.
Our data is in different tables. 

Before making some calculation, we should merge data from different Dataframes (merging in pandas is the same to joining in SQL).

```python
# Merge tables 'contracts' and 'locations' by the key 'loc_id'. 
contract_loc = pd.merge(contracts, locations, on = 'loc_id',how = 'left')
print(contract_loc.shape)
contract_loc.head()
```

```python
# Merge tables 'contract_loc' and 'cryield'
yields = pd.merge(cryield, contract_loc, on = 'cntr_id',how = 'left')
print(yields.shape)
yields.head()
```

For understanding crop yields from region to region, from district to district, from farmer to farmer. we should group data by those attributes separately for 'kg' and 'acre' and after get the metric kg/acre. 

```python
# Group by region/district/farmer and get acres of the crop by each attribute
yields.groupby(['reg_nm', 'dist_nm', 'cust_id_x'])['planted_acres'].sum()
```

```python
# Group by region/district/farmer and get a kg of the crop by each attribute
yields['kg'] = yields['yield_bags']*100
yields.groupby(['reg_nm', 'dist_nm', 'cust_id_x'])['kg'].sum()
```

##### 3.1.1.  Get the metric kg/acres for each region. Define the most efficient region

Define how many acres we have in each region in and display those values in %.

```python
acres_by_region = yields.groupby(['reg_nm'])['planted_acres'].sum()
print(acres_by_region)
plt.pie(acres_by_region.values,labels=acres_by_region.index, autopct='%1.1f%%')
plt.show()
```

We see that region 'GH-NP' has more than 62% of acres and 'GH-TV' does not have acres for crops at all. 

Define how many kg of crops we have in each region in and display those values in %.

```python
kg_by_region = yields.groupby(['reg_nm'])['kg'].sum()
print(kg_by_region)
plt.pie(kg_by_region.values,labels=kg_by_region.index, autopct='%1.1f%%')
plt.show()
```

The situation with kg of crops is similar to acres distribution between regions.

As the next step, we create the DataFrame where we display acres and kg for each region and as a final step, get the metric: kg/acres.

```python
# Display data in the DataFrame
data = {'acres':acres_by_region.values, 'kg':kg_by_region.values}
kg_acres_metric = pd.DataFrame(data=data, index = kg_by_region.index)
kg_acres_metric
```

```python
# Get the metric 
kg_acres_metric['kg_acres'] = kg_acres_metric['kg'] / kg_acres_metric['acres']
kg_acres_metric
```

##### Conclusion: despite GH-NP has higher values of acres and kg, the metric tells us that acres are used the best in regions: 'GH-UW' and 'GH-BA'. 'GH-NP' gets third place only.

'Ejura-Sekyedumase' and  'Savelugu-Nanton' give 18.3 and 17.2 % crops in the country.

##### 3.1.4. What seems to be a normal, vs. low, vs. high yield for maize in Ghana?

We use quantile to define those values. Everything between quantiles 0.25 and 0.75 will be normal. More than 0.75 quantiles will be a good result and below 0.25 will be a bad result. We also define high and low value for each customer.

```python
print('Normal yields is {} kg/acres. If we got yields between {} and {} we get average result'.format(kg_acres_metric['kg_acres'].quantile(0.5),kg_acres_metric['kg_acres'].quantile(0.25), kg_acres_metric['kg_acres'].quantile(0.75)))
print('The higher values is: {} kg/acres'.format(kg_acres_metric['kg_acres'].max()))
print('The lower values is: {} kg/acres'.format(kg_acres_metric['kg_acres'].min()))
```

Based on the lower value, we can conclude that it was customers who had acres but get 0 results. Probably, they did not seed.

We can also display it on the boxplot for a better understanding of data distribution.

```python
ax = sns.boxplot(x=kg_acres_metric['kg_acres'])
```

##### 3.4. At the district level, can we say that districts with poorer (or better) harvests received higher (or lower) insurance payout amounts?

Merge table 'influence' and 'contracts' to get 'kg_acres' and 'location_id' in one table.

```python
# Merging data
ins_payout = pd.merge(influence, contracts, on='cust_id', how='left')
print(ins_payout.shape)
ins_payout.head()
```

Merge 'ins_payout' and 'locations' to get 'dist_id' and 'payout' and 'kg_acres' in one table.

```python
dist_payout_metric = pd.merge(ins_payout, locations, on='loc_id', how='left')
dist_payout_metric.shape
```

As the next steps, we should group 'kg_acres' and 'payout' by 'district'.

```python
# Metric in each district
dist_kg_acres = dist_payout_metric.groupby('dist_id')['kg_acres'].mean()
# Payout in each district
dist_payout = dist_payout_metric.groupby('dist_id')['payout'].mean()
d = {'kg_acres':dist_kg_acres.values, 'payout':dist_payout.values}
# Get the final DataFrame
correlation = pd.DataFrame(data=d, index = dist_payout.index)
correlation.head()
```

As a final step, define the correlation between 'payout' and our metric.

```python
corrmat = correlation.corr()
corrmat
```

##### Conclusion: We get more payouts in regions with better yield.
